```python
def ph_evaluator(c=0.001):
  'Ph decision maker'
  ph_value= ph(c)

  if ph_value < 7:
    decision='kiselo'
  else:
    decision= 'lužnato'


  return decision
ph_evaluator()
```

```python
def ph_evaluator(c=0.001):

  if ph(c) < 7:
    return 'kiselo'

  else:
    return 'lužnato'
conc_list= [0.001, 0.01, 0.1, 1, 10]
for solution in conc_list:
  print(ph_evaluator(solution))
```

```python
def test_if_else(a,b):

  '''a,b su argumenti
  return 'a > b' of 'a in not > b'
  '''
  if a > b:
    print(a,'>', b)

  else:
    print('%d is not > %d' %(a,b))
```

```python
test_if_else(6,2)
```

```python
test_if_else(2,6)
```

```python
help(test_if_else)
```
